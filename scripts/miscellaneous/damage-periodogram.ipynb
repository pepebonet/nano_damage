```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cis = pd.read_csv('./rel_increase_cisplatin.tsv', sep='\t')
mms = pd.read_csv('./rel_increase_mms.tsv', sep='\t')
```

```python
signal_cis = cis['Relative Increase'].values
signal_mms = mms['Relative Increase'].values
```

```python
fig, ax = plt.subplots(figsize=(15, 3))
ax.plot(range(len(signal_cis)), signal_cis, label='cisplatin')
ax.plot(range(len(signal_mms)), signal_mms, label='MMS')
plt.legend()
plt.show()
```
